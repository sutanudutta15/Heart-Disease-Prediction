```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score,confusion_matrix
dataset = pd.read_csv("heart.csv")
print(dataset.isnull().sum())
data_dup = dataset.duplicated().any()
print(dataset.shape)
print(data_dup)
new_dataset = dataset.drop_duplicates()
print(new_dataset.duplicated().any())
print(new_dataset.shape)
```

```python
new_dataset.info()
```

```python
X = new_dataset.iloc[:, :-1].values
y = new_dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_classifier = LogisticRegression()
log_classifier.fit(X_train, y_train)
y_pred_log = log_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_log))
print(confusion_matrix(y_test,y_pred_log))
print(accuracy_score(y_train,log_classifier.predict(X_train)))
print(confusion_matrix(y_train,log_classifier.predict(X_train)))
```

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred_nb = nb_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_nb))
print(confusion_matrix(y_test,y_pred_nb))
print(accuracy_score(y_train,nb_classifier.predict(X_train)))
print(confusion_matrix(y_train,nb_classifier.predict(X_train)))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree_classifier.fit(X_train, y_train)
y_pred_tree = tree_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_tree))
print(confusion_matrix(y_test,y_pred_tree))
print(accuracy_score(y_train,tree_classifier.predict(X_train)))
print(confusion_matrix(y_train,tree_classifier.predict(X_train)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=10,criterion='entropy', random_state=0)
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_rf))
print(confusion_matrix(y_test,y_pred_rf))
print(accuracy_score(y_train,rf_classifier.predict(X_train)))
print(confusion_matrix(y_train,rf_classifier.predict(X_train)))
```

```python
from sklearn.svm import SVC
svc_rbf_classifier = SVC(kernel='rbf',random_state=0)
svc_rbf_classifier.fit(X_train, y_train)
y_pred_svc = svc_rbf_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_svc))
print(confusion_matrix(y_test,y_pred_svc))
print(accuracy_score(y_train,svc_rbf_classifier.predict(X_train)))
print(confusion_matrix(y_train,svc_rbf_classifier.predict(X_train)))
```

```python
from sklearn.svm import SVC
svc_linear_classifier = SVC(kernel='linear',random_state=0)
svc_linear_classifier.fit(X_train, y_train)
y_pred_svc_linear = svc_linear_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_svc_linear))
print(confusion_matrix(y_test,y_pred_svc_linear))
print(accuracy_score(y_train,svc_linear_classifier.predict(X_train)))
print(confusion_matrix(y_train,svc_linear_classifier.predict(X_train)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn_classifier.fit(X_train, y_train)
y_pred_knn = knn_classifier.predict(X_test)
print(accuracy_score(y_test,y_pred_knn))
print(confusion_matrix(y_test,y_pred_knn))
print(accuracy_score(y_train,knn_classifier.predict(X_train)))
print(confusion_matrix(y_train,knn_classifier.predict(X_train)))
```
